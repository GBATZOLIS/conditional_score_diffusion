```python
import json
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open('mammoth.json', 'rb') as f:
    mammoth_list = json.load(f)
```

```python
mammoth = np.array(mammoth_list)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.view_init(0, 35)
ax.scatter(mammoth[:,0], mammoth[:,1], mammoth[:,2])
fig.show()
```

```python
mammoth = mammoth - mammoth.mean() 
mammoth = mammoth / (mammoth.max() - mammoth.min())
mammoth
```

```python
from scipy.spatial.distance import cdist
distances = cdist(mammoth, mammoth)
```
