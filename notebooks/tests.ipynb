```python
import utils
import json
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open('mammoth.json', 'rb') as f:
    mammoth_list = json.load(f)
```

```python
mammoth = np.array(mammoth_list)
mammoth = mammoth - mammoth.mean(0) 
mammoth = mammoth / (mammoth.max() - mammoth.min())
```

```python
mammoth.max()
```

```python
plotting_n = 20000
data = mammoth
ids = np.random.randint(0, data.shape[0], plotting_n)
data = data[ids]
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.view_init(25, -20)
#ax.view_init(0, 0)
ax.scatter(data[:,0], data[:,1], data[:,2], alpha=0.1, color='g')
#ax.axis('off')
fig.savefig('figures/mammoth', dpi=300, facecolor='white')
fig.show()
```

```python
from scipy.spatial.distance import cdist
distances = cdist(mammoth, mammoth)
```
